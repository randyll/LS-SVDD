```python
import numpy as np
from sklearn.svm import SVR
import matplotlib.pyplot as plt

# Generating some sample data
np.random.seed(42)
X = np.sort(5 * np.random.rand(80, 1), axis=0)
y = np.sin(X).ravel()
y[::5] += 3 * (0.5 - np.random.rand(16))

# Create a SVR model
epsilon = 0.1  # The regularization parameter (epsilon-insensitive loss)
svr = SVR(kernel='rbf', epsilon=epsilon)

# Fit the model to the data
svr.fit(X, y)

# Generate predictions
X_pred = np.linspace(0, 5, 100).reshape(-1, 1)
y_pred = svr.predict(X_pred)

# Plot the results
plt.scatter(X, y, color='darkorange', label='Data')
plt.plot(X_pred, y_pred, color='navy', lw=2, label='LS-SVR Model')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Least Squares Support Vector Regression')
plt.legend()
plt.show()
```

```python
import numpy as np

class LSSVR:
    def __init__(self, C=1.0, epsilon=0.1, kernel='linear', sigma=1.0):
        self.C = C
        self.epsilon = epsilon
        self.kernel = kernel
        self.sigma = sigma
        self.support_vectors = None
        self.support_vector_labels = None

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.support_vectors = []
        self.support_vector_labels = []

        # Compute the Gram matrix
        gram_matrix = np.zeros((n_samples, n_samples))
        if self.kernel == 'linear':
            gram_matrix = np.dot(X, X.T)
        elif self.kernel == 'rbf':
            for i in range(n_samples):
                for j in range(n_samples):
                    diff = X[i] - X[j]
                    gram_matrix[i, j] = np.exp(-np.dot(diff, diff.T) / (2 * self.sigma ** 2))
        else:
            raise ValueError("Invalid kernel type. Only 'linear' and 'rbf' kernels are supported.")

        # Solve the dual optimization problem to find the Lagrange multipliers
        P = np.outer(y, y) * gram_matrix
        q = -np.ones(n_samples)
        G = np.vstack((-np.eye(n_samples), np.eye(n_samples)))
        h = np.hstack((np.zeros(n_samples), self.C * np.ones(n_samples)))
        A = y.reshape(1, -1)
        b = np.array([0.0])

        from cvxopt import matrix, solvers
        solvers.options['show_progress'] = False
        solution = solvers.qp(matrix(P), matrix(q), matrix(G), matrix(h), matrix(A), matrix(b))
        alpha = np.array(solution['x'])

        # Identify support vectors
        sv_indices = np.where(alpha > 1e-5)[0]
        self.support_vectors = X[sv_indices]
        self.support_vector_labels = y[sv_indices]
        self.alpha = alpha[sv_indices]

        # Calculate the intercept
        sv_index = sv_indices[0]
        self.intercept = y[sv_index] - np.sum(self.alpha * self.support_vector_labels * gram_matrix[sv_index, sv_indices])

    def predict(self, X):
        n_samples = X.shape[0]
        y_pred = np.zeros(n_samples)

        for i in range(n_samples):
            if self.kernel == 'linear':
                y_pred[i] = np.dot(self.alpha * self.support_vector_labels, np.dot(self.support_vectors, X[i]))
            elif self.kernel == 'rbf':
                for j in range(len(self.alpha)):
                    diff = X[i] - self.support_vectors[j]
                    y_pred[i] += self.alpha[j] * self.support_vector_labels[j] * np.exp(-np.dot(diff, diff.T) / (2 * self.sigma ** 2))
            else:
                raise ValueError("Invalid kernel type. Only 'linear' and 'rbf' kernels are supported.")
        
        return y_pred + self.intercept

# Example usage:
if __name__ == '__main__':
    import matplotlib.pyplot as plt

    # Generating some sample data
    np.random.seed(868)
    X = np.sort(5 * np.random.rand(80, 1), axis=0)
    y = np.sin(X).ravel()
    y[::5] += 3 * (0.5 - np.random.rand(16))

    
    # Create a LSSVR model
    ls_svr = LSSVR(C=1.0, epsilon=0.1, kernel='rbf')

    # Fit the model to the data
    ls_svr.fit(X, y)

    # Generate predictions
    X_pred = np.linspace(0, 5, 100).reshape(-1, 1)
    y_pred = ls_svr.predict(X_pred)

    # Plot the results
    plt.scatter(X, y, color='darkorange', label='Data')
    plt.plot(X_pred, y_pred+39, color='navy', lw=2, label='LS-SVR Model')
    plt.xlabel('X')
    plt.ylabel('y')
    plt.title('Least Squares Support Vector Regression')
    plt.legend()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

class LSSVDD:
    def __init__(self, C=1.0, epsilon=0.1, kernel='linear', sigma=1.0):
        self.C = C
        self.epsilon = epsilon
        self.kernel = kernel
        self.sigma = sigma
        self.support_vectors = None
        self.support_vector_labels = None

    def fit(self, X):
        n_samples, n_features = X.shape
        self.support_vectors = X

        # Compute the Gram matrix
        gram_matrix = np.zeros((n_samples, n_samples))
        if self.kernel == 'linear':
            gram_matrix = np.dot(X, X.T)
        elif self.kernel == 'rbf':
            for i in range(n_samples):
                for j in range(n_samples):
                    diff = X[i] - X[j]
                    gram_matrix[i, j] = np.exp(-np.dot(diff, diff.T) / (2 * self.sigma ** 2))
        else:
            raise ValueError("Invalid kernel type. Only 'linear' and 'rbf' kernels are supported.")

        # Solve the dual optimization problem using the closed-form solution
        P = np.outer(np.ones(n_samples), np.ones(n_samples))
        q = -np.ones(n_samples)
        G = np.vstack((-np.eye(n_samples), np.eye(n_samples)))
        h = np.hstack((np.zeros(n_samples), self.C * np.ones(n_samples)))

        A = np.diag(np.ones(n_samples))
        b = np.zeros(n_samples)

        # Calculate the Lagrange multipliers using the closed-form solution
        from scipy.linalg import solve
        lagrange_multipliers = solve(np.dot(P, gram_matrix), q,  overwrite_a=True, check_finite=False)

        # Identify support vectors
        sv_indices = np.where(np.abs(lagrange_multipliers) > 1e-5)[0]
        self.support_vector_labels = lagrange_multipliers[sv_indices]

    def predict(self, X):
        n_samples = X.shape[0]
        y_pred = np.zeros(n_samples)

        for i in range(n_samples):
            if self.kernel == 'linear':
                y_pred[i] = np.dot(self.support_vector_labels, np.dot(self.support_vectors, X[i]))
            elif self.kernel == 'rbf':
                for j in range(len(self.support_vector_labels)):
                    diff = X[i] - self.support_vectors[j]
                    y_pred[i] += self.support_vector_labels[j] * np.exp(-np.dot(diff, diff.T) / (2 * self.sigma ** 2))
            else:
                raise ValueError("Invalid kernel type. Only 'linear' and 'rbf' kernels are supported.")

        return y_pred

# Example usage:
if __name__ == '__main__':
    np.random.seed(42)
    normal_data = np.random.randn(100, 2)
    anomaly_data = np.random.randn(20, 2) + 4.0

    X = np.vstack((normal_data, anomaly_data))

    ls_svdd = LSSVDD(C=1.0, epsilon=0.1, kernel='rbf')

    ls_svdd.fit(X)

    X_pred = np.linspace(-2, 6, 100)
    Y_pred = np.linspace(-2, 6, 100)

    X_pred, Y_pred = np.meshgrid(X_pred, Y_pred)
    Z_pred = np.zeros(X_pred.shape)

    for i in range(X_pred.shape[0]):
        for j in range(X_pred.shape[1]):
            Z_pred[i, j] = ls_svdd.predict(np.array([X_pred[i, j], Y_pred[i, j]]))

    plt.contourf(X_pred, Y_pred, Z_pred, levels=np.linspace(np.min(Z_pred), np.max(Z_pred), 10), cmap=plt.cm.PuBu)
    plt.scatter(X[:, 0], X[:, 1], color='darkorange')
    plt.title('LS-SVDD Anomaly Detection')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist

class LSSVDD:
    def __init__(self, C=1.0, epsilon=0.1, kernel='linear', sigma=1.0):
        self.C = C
        self.epsilon = epsilon
        self.kernel = kernel
        self.sigma = sigma
        self.support_vectors = None
        self.support_vector_labels = None

    def fit(self, X):
        n_samples = X.shape[0]
        self.support_vectors = X

        # Compute the Gram matrix
        gram_matrix = self.compute_gram_matrix(X)

        # Solve the dual optimization problem using the closed-form solution
        P = np.outer(np.ones(n_samples), np.ones(n_samples))
        q = -np.ones(n_samples)
        G = np.vstack((-np.eye(n_samples), np.eye(n_samples)))
        h = np.hstack((np.zeros(n_samples), self.C * np.ones(n_samples)))

        A = np.diag(np.ones(n_samples))
        b = np.zeros(n_samples)

        # Calculate the Lagrange multipliers using the closed-form solution
        from scipy.linalg import solve
        lagrange_multipliers = solve(np.dot(P, gram_matrix), q, overwrite_a=True, check_finite=False)

        # Identify support vectors
        sv_indices = np.where(np.abs(lagrange_multipliers) > 1e-5)[0]
        self.support_vector_labels = lagrange_multipliers[sv_indices]

    def compute_gram_matrix(self, X):
        n_samples = X.shape[0]
        gram_matrix = np.zeros((n_samples, n_samples))

        if self.kernel == 'linear':
            gram_matrix = np.dot(X, X.T)
        elif self.kernel == 'rbf':
            pairwise_dists = cdist(X, X, metric='euclidean')
            gram_matrix = np.exp(-pairwise_dists ** 2 / (2 * self.sigma ** 2))
        else:
            raise ValueError("Invalid kernel type. Only 'linear' and 'rbf' kernels are supported.")

        return gram_matrix

    def predict(self, X):
        n_samples = X.shape[0]
        y_pred = np.zeros(n_samples)

        for i in range(n_samples):
            if self.kernel == 'linear':
                y_pred[i] = np.dot(self.support_vector_labels, np.dot(self.support_vectors, X[i]))
            elif self.kernel == 'rbf':
                pairwise_dists = cdist(X[i].reshape(1, -1), self.support_vectors, metric='euclidean')
                y_pred[i] = np.sum(self.support_vector_labels * np.exp(-pairwise_dists ** 2 / (2 * self.sigma ** 2)))
            else:
                raise ValueError("Invalid kernel type. Only 'linear' and 'rbf' kernels are supported.")

        return y_pred

# Example usage:
if __name__ == '__main__':
    np.random.seed(42)
    normal_data = np.random.randn(100, 2)
    anomaly_data = np.random.randn(20, 2) + 4.0

    X = np.vstack((normal_data, anomaly_data))

    ls_svdd = LSSVDD(C=1.0, epsilon=0.1, kernel='rbf', sigma=1.0)

    ls_svdd.fit(X)

    X_pred = np.linspace(-2, 6, 100)
    Y_pred = np.linspace(-2, 6, 100)

    X_pred, Y_pred = np.meshgrid(X_pred, Y_pred)
    Z_pred = np.zeros(X_pred.shape)

    for i in range(X_pred.shape[0]):
        for j in range(X_pred.shape[1]):
            Z_pred[i, j] = ls_svdd.predict(np.array([X_pred[i, j], Y_pred[i, j]]))

    plt.contourf(X_pred, Y_pred, Z_pred, levels=np.linspace(np.min(Z_pred), np.max(Z_pred), 10), cmap=plt.cm.PuBu)
    plt.scatter(X[:, 0], X[:, 1], color='darkorange')
    plt.title('LS-SVDD Anomaly Detection')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.show()
```
